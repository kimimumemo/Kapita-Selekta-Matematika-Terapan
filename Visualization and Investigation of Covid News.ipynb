## Soal 1

```python
from bs4 import BeautifulSoup

text_scrap = ''
with open ('data.html','r') as html_file:
  # membaca file
  content = html_file.read()

  # jadiin baris
  soup = BeautifulSoup(content, 'lxml')
  # memberikan definisi untuk setiap tag
  course_html_tags = soup.find_all('div', class_='ants-news-section ants-news-content min-height:300px;')
  for course in course_html_tags:
    # print(course)
    text_scrap += str(course) + '\n'

print(text_scrap)
```

**Preprocessing**

```python
import nltk
nltk.download('punkt')
```

```python
import string 
import re #regex library
from nltk.tokenize import word_tokenize 
```

```python
# mengubah huruf-huruf yang ada pada kalimat supaya menjadi berhuruf kecil
Lower_string = text_scrap.lower()
```

```python
Lower_string
```

```python
# menghapus bilangan bulat (integer) pada kalimat 
Lower_string = re.sub(r"\d+", "", Lower_string)
```

```python
Lower_string
```

```python
#remove punctuation
Lower_string = re.sub(r'[^\w\s]', ' ', Lower_string)
```

```python
Lower_string
```

```python
#remove whitespace leading & trailing
Lower_string = Lower_string.strip()
```

```python
Lower_string
```

```python
#remove multiple whitespace into single whitespace
Lower_string = re.sub('\s+',' ',Lower_string)
```

```python
Lower_string
```

```python
tokens = nltk.tokenize.word_tokenize(Lower_string)

print('Tokenizing Result : \n') 
print(tokens)
```

```python
# import word_tokenize & FreqDist from NLTK
from nltk.tokenize import word_tokenize 
from nltk.probability import FreqDist
```

```python
freq_tokens = nltk.FreqDist(tokens)

print('Frequency Tokens : \n') 
print(freq_tokens.most_common())
```

```python
nltk.download('stopwords')
```

```python
# menghapus kata-kata yang tidak diperlukan
# mendefinisikan kata-kata hasil transfer dari html bernama unnecessary_words
# mendefinisikan kata-kata berjenis stopwords bernama list_stopwords
from nltk.corpus import stopwords
list_stopwords = set(stopwords.words('indonesian'))

# terlampir pada halaman terakhir tabel unnecessary_words yang saya gunakan
unnecessary_words = ['div', 'class', 'ants', 'news', 'section', 'ants', 'news', 'content', 'min', 'height', 'px', 'ul', 'li', 'style', 'padding', 'contain', 'px', 'h', 'per', 'container', 'title', 'style', 'margin', 'bottom','h', 'span', 'class', 'date', 'style', 'display', 'block', 'strongdot', 'id', 'd', 'br', 'span', 'strong', 'web', 'assets', 'title', 'png', 'article', 'halo', 'img', 'go', 'at', 'dot', 'src', 'fl', 'ext', 'images', 'image', 'per', 'ui', 'ni', 'jpeg']
result = []

for i in range(0,len(tokens)):
  new_tokens = tokens[i]
  if new_tokens not in unnecessary_words:
    if new_tokens not in list_stopwords:
      result.append(new_tokens)
```

```python
freq_result = nltk.FreqDist(result)

print('Frequency Tokens : \n') 
print(freq_result.most_common())
```

```python
import pandas as pd

df_freq_result = pd.DataFrame.from_dict(freq_result, orient='index')
df_freq_result.columns = ['Frequency']
df_freq_result.index.name = 'Key'

df_freq_result.plot(kind='bar')
```

Stemming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
stemmer = PorterStemmer()
```

```python
# terlihat pada tahap ini bahwa token merupakan kata-kata penting
# yang dibutuhnkan untuk melihat kecenderungan artikel tersebut

# token without stopword
list_tokens = result

# stem
output = [(token + " : " + stemmer.stem(token)) for token in list_tokens]

output
```

```python
# hasil dari stemming tidak saya gunakan karena kurang baik melalukan pemotongan
# pada kata awal
text = ''
for t in result:
  text += str(t) + ' '
```

```python
# berikut merupakan word cloud yang lebih efektif dalam penggunaan penghapusan dan
# pembersihan dari artikel tersebut
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS 
# stopwords is a collection of words that dont convey meaning. mostly pronouns such as he she etc.
wordcloud = WordCloud(width = 600, height = 300, random_state=1, background_color='white', collocations=False, stopwords = STOPWORDS).generate(text)
plt.figure(figsize=(40, 30))
# Display image
plt.imshow(wordcloud) 
# No axis 
plt.axis("off")
plt.show()
```
